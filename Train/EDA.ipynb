```python
import pickle
from typing import List
def open_pkl(filename: str) -> List: 
    f = open(filename, 'rb')
    elementList = []
    while (True):
        try:
            data = pickle.load(f)
            elementList.append(data)
        except EOFError:
            return elementList
```
